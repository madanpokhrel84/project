```python
from bs4 import BeautifulSoup
import requests
```

```python
url = "https://en.wikipedia.org/wiki/List_of_largest_companies_in_the_United_States_by_revenue"
page = requests.get(url)
soup = BeautifulSoup(page.text, "lxml")
soup
```

```python
table = soup.find('table', class_ = "wikitable sortable")
table
```

```python
table_heads = table.find_all('th')
table_heads
```

```python
table_heads = [head.text.strip() for head in table_heads]
print(table_heads)
```

```python
table_rows = table.find_all('tr')
table_rows
```

```python
row_data = [each_row.text.strip() for each_row in table_rows]
row_data
```

```python
stripped_data = [data.split("\n\n") for data in row_data]
stripped_data[1:]
```

```python
import pandas as pd
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
```

```python
df = pd.DataFrame(stripped_data[1:], columns= table_heads)
df
```

```python
df["Revenue (USD millions)"] = list(map(lambda x: int(x.replace(",","")), list(df["Revenue (USD millions)"])))
df["Revenue growth"] = list(map(lambda x: float(x.replace("%","")), list(df["Revenue growth"])))
df["Employees"] = list(map(lambda x: int(x.replace(",","")), list(df["Employees"])))
df
```

```python
import csv
df.to_csv(r"C:\Users\yubra\Jupyter Notebook\python_selfStudy\Companies.csv",index=False, mode='w', header=True)
```

```python
df_sorted = df.sort_values(by='Revenue growth', ascending=False)
top_10_companies = df_sorted.head(10)
import plotly.express as px
fig = px.pie(
    top_10_companies,
    values='Revenue growth',
    names='Name',
    title='Top 10 Companies with Highest Revenue Growth'
)

fig.show()
```
